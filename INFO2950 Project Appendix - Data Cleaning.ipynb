### 3. Clean Up Columns and Rows <br>

For the subscribers and revenues dataset, we eliminated the rows and columns in which the values were null. We also filtered out the columns so that we only have the columns that we need which are: number of subscribers by quarters and company revenue by quarters. Furthermore, we renamed the columns so that they are more easily readable during analysis.

```python
#Eliminate columns
df_subscribers_and_revenue.drop(columns=['Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33',
       'Unnamed: 34',' Q1 2018 Average Monthly Revenue per Membership ', 
                    ' Q2 2018 Average Monthly Revenue per Membership ', 
                    'Q3 2018 Average Monthly Revenue per Membership',
                    'Q4 2018 Average Monthly Revenue per Membership',
                    'Q1 2019 Average Monthly Revenue per Membership',
                    'Q2 2019 Average Monthly Revenue per Membership',
                    'Q3 2019 Average Monthly Revenue per Membership',
                    'Q4 2019 Average Monthly Revenue per Membership',
                    'Q1 2020 Average Monthly Revenue per Membership',
                    'Q2 2020 Average Monthly Revenue per Membership'], axis=1, inplace=True)

#Eliminate rows
df_subscribers_and_revenue.drop([4,5,6,7,8,9,10], inplace=True)

#Rename Columns
df_subscribers_and_revenue = df_subscribers_and_revenue.rename(columns={
    'Q1 2018 # of Subscribers': 'Q1 2018 Subscriber','Q1 2018 Quarter Revenue':"Q1 2018 Revenue",
    'Q2 2018 # of Subscribers': 'Q2 2018 Subscriber','Q2 2018 Quarter Revenue':"Q2 2018 Revenue",
    'Q3 2018 # of Subscribers': 'Q3 2018 Subscriber','Q3 2018 Quarter Revenue':"Q3 2018 Revenue",
    'Q4 2018 # of Subscribers': 'Q4 2018 Subscriber','Q4 2018 Quarter Revenue':"Q4 2018 Revenue",
    'Q1 2019 # of Subscribers': 'Q1 2019 Subscriber','Q1 2019 Quarter Revenue':"Q1 2019 Revenue",
    'Q2 2019 # of Subscribers': 'Q2 2019 Subscriber','Q2 2019 Quarter Revenue':"Q2 2019 Revenue",
    'Q3 2019 # of Subscribers': 'Q3 2019 Subscriber','Q3 2019 Quarter Revenue':"Q3 2019 Revenue",
    'Q4 2019 # of Subscribers': 'Q4 2019 Subscriber','Q4 2019 Quarter Revenue':"Q4 2019 Revenue",
    'Q1 2020 # of Subscribers': 'Q1 2020 Subscriber','Q1 2020 Quarter Revenue':"Q1 2020 Revenue",
    'Q2 2020 # of Subscribers': 'Q2 2020 Subscriber','Q2 2020 Quarter Revenue':"Q2 2020 Revenue"
})
```

```python
df_subscribers_and_revenue.head(3)
```

For the USA daily covid-19 dataset, we are only interested in the total/increase of tests and positive cases. Thus, we have eliminated other columns including negative cases, pending cases, hospitalized cases, and etc.

```python
#Eliminate columns
df_covid = df_covid[['date','positive','totalTestResults','positiveIncrease','totalTestResultsIncrease']]
```

```python
df_covid.head()
```

For the mobility dataset, we selected only the rows that correspond to nationwide statistics because we will not be focusing on sub-region data (we only need the national mobility tendency from before the pandemic). In the same vein, we eliminated the columns that indicate sub-region data. We also renamed columns so that they are more readable.

```python
#Keeping only nationwide data
df_mobility = df_mobility[df_mobility['sub_region_1'].isnull()]

#Eliminate columns
df_mobility.drop(columns=['country_region_code','sub_region_1', 'sub_region_2',
       'metro_area', 'iso_3166_2_code', 'census_fips_code'], inplace=True)

#Rename Columns
df_mobility = df_mobility.rename(columns={
    'retail_and_recreation_percent_change_from_baseline': 'retail_and_recreation',
    'grocery_and_pharmacy_percent_change_from_baseline': 'grocery_and_pharmacy',
    'parks_percent_change_from_baseline': 'parks',
    'transit_stations_percent_change_from_baseline': 'transit_station',
    'workplaces_percent_change_from_baseline': 'workplace',
    'residential_percent_change_from_baseline': 'residence'
})
```

```python
df_mobility.head()
```

For the screen time datasets before and after the coronavirus, we eliminated the columns that contained the headcounts of the respondents. We only need the percentage data of the respondents. Also, we only kept the first row 'Parents' since the rows below were merely 'Parents' divided into demographic groups. We also provided more

```python
#Eliminate columns
df_screentime_before_corona.drop(columns=['Demographic','Unnamed: 2', 'Unnamed: 4','Unnamed: 6','Unnamed: 8',
       'Unnamed: 10', 'Total N', 'Unnamed: 13'], axis=1, inplace=True)
df_screentime_after_corona.drop(columns=['Demographic','Unnamed: 2', 'Unnamed: 4','Unnamed: 6','Unnamed: 8',
       'Unnamed: 10', 'Total N', 'Unnamed: 13'], axis=1, inplace=True)

#Eliminate rows
df_screentime_before_corona = df_screentime_before_corona.head(1)
df_screentime_after_corona = df_screentime_after_corona.head(1)

# #Change column names
df_screentime_before_corona = df_screentime_before_corona.rename(columns={'hours':'9+ hours'})
df_screentime_after_corona = df_screentime_after_corona.rename(columns={'hours':'9+ hours'})
```

```python
df_screentime_before_corona.head()
```

```python
df_screentime_after_corona.head()
```

We concatenated screentime datasets before and after the pandemic in order to facilitate the comparison

```python
#Concatenation & resetting index
df_screentime = pd.concat([df_screentime_before_corona,df_screentime_after_corona])
df_screentime = df_screentime.reset_index()
df_screentime.drop(columns=['index'], inplace=True)
```

We divided the netflix subscribers and revenue dataset into two separate datasets because we will analyze subscribers data and revenue data separately.

```python
#Separate subscribers data
df_subscribers = df_subscribers_and_revenue[['Region', 
                                             'Q1 2018 Subscriber',
                                             'Q2 2018 Subscriber',
                                             'Q3 2018 Subscriber',
                                             'Q4 2018 Subscriber',
                                             'Q1 2019 Subscriber',
                                             'Q2 2019 Subscriber',
                                             'Q3 2019 Subscriber',
                                             'Q4 2019 Subscriber',
                                             'Q1 2020 Subscriber',
                                             'Q2 2020 Subscriber']]

#Separate revenues data
df_revenue = df_subscribers_and_revenue[['Region',
                                             'Q1 2018 Revenue',
                                             'Q2 2018 Revenue',
                                             'Q3 2018 Revenue',
                                             'Q4 2018 Revenue',
                                             'Q1 2019 Revenue',
                                             'Q2 2019 Revenue',
                                             'Q3 2019 Revenue',
                                             'Q4 2019 Revenue',
                                             'Q1 2020 Revenue',
                                             'Q2 2020 Revenue']]
```

### 6. Type Conversion<br><br>
Since the screen time data were strings with % attached in the end, I have eliminated all percentage symbols and changed the data type into integer in order to facilitate comparison and analysis.

```python
#Getting rid of symbols & Type Conversion
df_screentime = df_screentime.replace({'%':''}, regex=True)
df_screentime = df_screentime.astype('int32')
```
